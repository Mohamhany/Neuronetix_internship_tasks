## **define features and target**

```python
x=df.drop(columns=['income'])
```

```python
y=df[['income']]
y
```

## **data Scaling**

```python
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

## **Bulid Model**

```python
LR=LinearRegression()
LR.fit(x,y)
```

```python
LR.intercept_
```

```python
LR.coef_
```

```python
y_pred=LR.predict(x)
```

```python
print("accuracy:",int(r2_score(y,y_pred)*100),"%")
```
